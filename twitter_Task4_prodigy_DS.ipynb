```
train.columns = ['id', 'game', 'sentiment', 'text']
```

## **Remove Duplicates Data Cleaning**

```
train_cleaned= train.drop_duplicates()
test_cleaned = test.drop_duplicates()
```

## **Visualizations**

```
plt.figure(figsize=(10, 6))
sns.countplot(x='sentiment', data=train)
plt.title('Sentiment Distribution in Training Data')
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.countplot(x='sentiment', data=test)
plt.title('Sentiment Distribution in Testing Data')
plt.show()
```

```
positive_tweets_training = " ".join(tweet for tweet in train_cleaned[train_cleaned['Sentiment'] == 'Positive']['Text'])
wordcloud_positive_training = WordCloud(width=800, height=400, background_color='white').generate(positive_tweets_training)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_positive_training, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Positive Tweets in Training Data')
plt.show()
```

```
negative_tweets_training = " ".join(tweet for tweet in train_cleaned[train_cleaned['Sentiment'] == 'Negative']['Text']) # Replace 'Text' with the correct column name
wordcloud_negative_training = WordCloud(width=800, height=400, background_color='white').generate(negative_tweets_training)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_negative_training, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Negative Tweets in Training Data')
plt.show()
```

```
train['tweet_length'] = train['text'].astype(str).apply(len)
test['tweet_length'] = test['text'].astype(str).apply(len)
```

```
plt.figure(figsize=(10, 6))
sns.histplot(train, x='tweet_length', hue='sentiment', element='step', stat='density', common_norm=False)
plt.title('Tweet Length Distribution by Sentiment in Training Data')
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.histplot(test, x='tweet_length', hue='sentiment', element='step', stat='density', common_norm=False)
plt.title('Tweet Length Distribution by Sentiment in Testing Data')
plt.show()
```

```
sentiment_counts_training = train_cleaned['Sentiment'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(sentiment_counts_training, labels=sentiment_counts_training.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Sentiment Distribution in Training Data')
plt.axis('equal')
plt.show()
```

```
game_types = test['game'].value_counts()
game_types
```

```
plt.figure(figsize=(14,7))
palette = sns.color_palette("husl", len(game_types))
sns.barplot(x=game_types.values,y=game_types.index)
plt.title('Plot of Games and their count')
plt.ylabel('Type')
plt.xlabel('Count')

plt.show()
```
