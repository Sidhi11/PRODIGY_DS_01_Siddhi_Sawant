## **HISTOGRAM for yeach year population**

```python
cols=['2023 population','2022 population', '2020 population', '2015 population',
       '2010 population', '2000 population', '1990 population','1980 population', '1970 population']
```

```python
for i in cols:
        fig = plt.figure(figsize=(15,5))
        plt.hist(df[i],color='blue',bins=15)
        plt.title(i)
        plt.show()
```

## **BAR PLOT [Shows Top 5 Countries population for each year]**

```python
for col in cols:
    plt.figure(figsize=(12, 8))
    top_countries = df[['country', col]].sort_values(by=col, ascending=False).head(5)
    sns.barplot(x=col, y='country', data=top_countries, palette='viridis')
    plt.title(f'Top 5 Countries by {col}')
    plt.xlabel('Population')
    plt.ylabel('Country')
    plt.show()
```
